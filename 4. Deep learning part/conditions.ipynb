```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes_20240610.csv") 
raw['CRASH DATE'] = pd.to_datetime(raw['CRASH DATE'], format='%m/%d/%Y')
raw['YEAR'] = raw['CRASH DATE'].dt.year
raw = raw[(raw['BOROUGH'] == 'MANHATTAN') & (raw['YEAR'] == 2019)]
selected_columns = ['CRASH DATE', 'CRASH TIME']
conditions = raw[selected_columns]
```

```python
conditions
```

```python
import pandas as pd

# 假设 conditions 数据框已经被定义，并包含 CRASH DATE 和 CRASH TIME 列

# 确保 CRASH DATE 列被转换为 datetime 类型，先不指定格式以处理混合格式
conditions['CRASH DATE'] = pd.to_datetime(conditions['CRASH DATE'], errors='coerce')

# 处理转换失败的情况
if conditions['CRASH DATE'].isnull().any():
    print("Some dates could not be parsed.")

# 将 CRASH DATE 和 CRASH TIME 列转换为 datetime 格式并合并
conditions['CRASH DATETIME'] = pd.to_datetime(conditions['CRASH DATE'].astype(str) + ' ' + conditions['CRASH TIME'], errors='coerce')

# 处理转换失败的情况
if conditions['CRASH DATETIME'].isnull().any():
    print("Some datetimes could not be parsed.")

# 创建一个包含2019年所有日期和小时的完整时间范围
all_hours = pd.date_range(start='2019-01-01', end='2019-12-31 23:00:00', freq='H')

# 创建一个新的DataFrame，其中包含所有日期和小时
full_range_df = pd.DataFrame(all_hours, columns=['CRASH DATETIME'])

# 将原始数据与完整时间范围进行合并，以确保每个小时都有相应的数据
merged_df = full_range_df.merge(conditions, on='CRASH DATETIME', how='left')

# 按小时统计crash数量
hourly_crash_counts = merged_df.groupby(merged_df['CRASH DATETIME'].dt.floor('H')).size().reset_index(name='CRASH COUNT')

# 查看统计结果
hourly_crash_counts
```
