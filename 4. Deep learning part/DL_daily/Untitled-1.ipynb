```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# 读取数据
path = r"C:\Users\yanzh\Desktop\code_and_data\4. Deep learning part\处理数据\daily\MLP.csv"
data = pd.read_csv(path)

# 转换日期格式
data['date'] = pd.to_datetime(data['date'])

# 绘制 temperature_2m_max 的直方图和Q-Q图
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['temperature_2m_max'], kde=True)
plt.title('Histogram of temperature_2m_max')

plt.subplot(1, 2, 2)
stats.probplot(data['temperature_2m_max'], dist="norm", plot=plt)
plt.title('Q-Q plot of temperature_2m_max')

plt.show()

# 绘制 temperature_2m_min 的直方图和Q-Q图
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['temperature_2m_min'], kde=True)
plt.title('Histogram of temperature_2m_min')

plt.subplot(1, 2, 2)
stats.probplot(data['temperature_2m_min'], dist="norm", plot=plt)
plt.title('Q-Q plot of temperature_2m_min')

plt.show()

# 绘制 CRASH COUNT 的直方图和Q-Q图
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['CRASH COUNT'], kde=True)
plt.title('Histogram of CRASH COUNT')

plt.subplot(1, 2, 2)
stats.probplot(data['CRASH COUNT'], dist="norm", plot=plt)
plt.title('Q-Q plot of CRASH COUNT')

plt.show()

# 执行 Shapiro-Wilk 正态性检验
def perform_shapiro_test(column):
    stat, p_value = stats.shapiro(data[column])
    print(f'Shapiro-Wilk Test for {column}:\nStatistic={stat}, p-value={p_value}\n')

perform_shapiro_test('temperature_2m_max')
perform_shapiro_test('temperature_2m_min')
perform_shapiro_test('CRASH COUNT')
```
